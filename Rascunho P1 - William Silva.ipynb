### Sobre a OEA - Organização dos Estados Americanos
A OEA é uma organização internacional, que foi fundada em 1948, e que possui atualmente 35 membros independentes, pertencentes ao continente americano. Iniciou com 21 países, mas com o tempo foi conseguindo mais adeptos. Os países se comprometeram em defender os interesses do continente, visando a soluções pacíficas para o desenvolvimento, social, econômico e cultural. Ademais, tem objetivo de assegurar a democracia nos países e fortalecer a discussão acerca de tráfico de drogas, comércio, integração econômica, além de também prezar pelo fim da pobreza extrema e pela redução da desigualdade.
    

A partir disso, parto para minha análise, procurando saber se alguns dos pilares estão sendo cumpridos de fato. Para isso, passo pelo tópico do desenvolvimento associado à desigualdade. 

Para iniciar, pressuponho que, após se juntar à OEA, um país vai aumentar seu PIB Per Capita, tendo em vista que irá manter mais relações com os países do continente (e são mais de 30), o que aumenta suas trocas comerciais, consequentemente gerando mais riquezas para o país, impactando positivamente no PIB, independente da quantidade de pessoas que nascem. Por outro lado, acredito que isso aumentará o índice de gini, mesmo que o PIB Per Capita aumente, tendo em vista que medidas neoliberais, como a abertura comercial, sem o auxílio de propostas auxiliadoras, têm a tendência de causar maior desigualdade social no país, demonstrando uma possível concentração de riqueza.

# **Dados do ÍNDICE DE GINI**

Primeiramente, será feita a análise do Índice de Gini, que mede o nível de desigualdade dos países. Quanto maior o índice, mais desigual é aquele país, naquele determinado ano. A análise irá consistir em ver se os países que entraram à OEA apresentaram uma redução na desigualdade, para ver se de fato esse objetivo da OEA está sendo cumprido. A hipótese é de que não tenha redução significativa, mas tenha um aumento, tendo em vista a aplicação de políticas neoliberais.


```python
# Identificando quais são os países que pertencem à OEA
# Os país São Cristóvão e Neves e o país São Vicente e Granadinas não estão identificados nas
# outras tabelas, portanto não foram identificados.
gini_paises_da_oea = gini.loc[
            (gini.country=="Antigua and Barbuda") |
            (gini.country=="Argentina") |
            (gini.country=="Bahamas") | 
            (gini.country=="Barbados") |
            (gini.country=="Belize") |
            (gini.country=="Bolivia") |
            (gini.country=="Brazil") |
            (gini.country=="Canada") |
            (gini.country=="Chile") | 
            (gini.country=="Colombia") | 
            (gini.country=="Costa Rica") |
            (gini.country=="Cuba") |
            (gini.country=="Dominica") |
            (gini.country=="El Salvador") |
            (gini.country=="Ecuador") |
            (gini.country=="United States") |
            (gini.country=="Grenada") |
            (gini.country=="Guatemala") |
            (gini.country=="Guyana") |
            (gini.country=="Haiti") |
            (gini.country=="Honduras") |
            (gini.country=="Jamaica") |
            (gini.country=="Mexico") |
            (gini.country=="Nicaragua") | 
            (gini.country=="Panama") |
            (gini.country=="Paraguay") |
            (gini.country=="Dominican Republic") |
            (gini.country=="Saint Lucia") |
            (gini.country=="Suriname") | 
            (gini.country=="Trinidad and Tobago") | 
            (gini.country=="Uruguay") |
            (gini.country=="Venezuela")
]

# Mudando o "country" para "País" e deixando o índice como o nome do país
gini_new = gini_paises_da_oea.rename(columns = {'country':'Pais'})
gini_oea = gini_new.set_index("Pais")

# Foi colocado 220 para delitimar o período, tendo em vista que na tabela já possui previsões
gini_oea.iloc[:,:220]
```

Agora que possuímos dados de gini dos países, podemos separar um período de tempo para analisar seus dados. Como a OEA foi criada em 1948, é interessante ressaltar um período considerável antes e depois da data, que foi 1948.

Como há muitos dados para se vizualizar, fica interessante separar alguns países para se ter uma vizualização melhor e mais focada, para se ter uma ideia do geral.

O Canadá, por exemplo, só adentrou à OEA em 1990, sendo insustentável para ser considerado na análise desse período.

O Brasil, no entanto, pertence à organização desde seu início, assim como a Argentina e o Uruguai.

Por conta disso, é interessante comparar tais países, já que não possuem uma disparidade tão grande de perfil econômico. O Brasil têm maior PIB, no entanto não se distancia do perfil dos outros países, que também são economias emergentes.

Abaixo, observamos os dados dos três países, de 1920 a 1980.

```python
gini_oea.iloc[:, 120:181].loc[["Brazil", "Argentina", "Uruguay"]]
```

Para se ter uma melhor vizualização, é relevante colocar em um gráfico de dispersão, para vizualizar o crescimento ou o decrescimento do Índice de Gini.

***Importante lembrar que o ano de criação da OEA foi 1948***

```python
# Separando em três para obter seus dados
br = gini_oea.iloc[:, 120:181].loc[["Brazil"]]
ar = gini_oea.iloc[:, 120:181].loc[["Argentina"]]
ur = gini_oea.iloc[:, 120:181].loc[["Uruguay"]]

# Pegar os anos (colunas)
coluna = gini_oea.iloc[:, 120:181].loc[["Brazil", "Argentina", "Uruguay"]]
colunas = []
for i in coluna:
    colunas.append(i)

# Pegando os valores de gini do Brasil    
gini_brazil = []
for i in np.arange(0,len(colunas),1):
    gini_brazil.append(float(br[colunas[i]]))

# Pegando os valores de gini da Argentina
gini_argentina = []
for i in np.arange(0,len(colunas),1):
    gini_argentina.append(float(ar[colunas[i]]))
    
# Pegando os valores de gini do Uruguai
gini_uruguay = []
for i in np.arange(0,len(colunas),1):
    gini_uruguay.append(float(ur[colunas[i]]))
    
# Plotando o gráfico da evolução do Índice de Gini no decorrer dos anos
plt.scatter(colunas, gini_brazil, c='green', marker='.', label='Brazil')
plt.scatter(colunas, gini_argentina, c='blue', marker='.', label='Argentina')
plt.scatter(colunas, gini_uruguay, c='yellow', marker='.', label='Uruguay')
plt.title("Evolução do Índice de Gini")
plt.xlabel("Tempo [Anos]")
plt.ylabel("Índice de Gini [%]")
plt.legend()
plt.grid(True)
plt.show()
```

Para o Canadá, país que só entrou na OEA em 1990, têm-se:
